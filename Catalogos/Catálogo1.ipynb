```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen original
A = cv2.imread('imagen1.jpg')
A = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)  # Convertir la imagen de BGR a RGB para mostrar correctamente con Matplotlib

# Mostrar la imagen original
plt.subplot(1, 3, 1)
plt.imshow(A)
plt.title('Imagen original')
plt.axis('off')

# Incorporar ruido sal y pimienta
def add_salt_and_pepper_noise(image, amount):
    noisy = image.copy()
    # Generar ruido
    num_salt = np.ceil(amount * image.size * 0.5)
    num_pepper = np.ceil(amount * image.size * 0.5)
    
    # A√±adir sal (pixeles blancos)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy[coords[0], coords[1], :] = 1
    
    # A√±adir pimienta (pixeles negros)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy[coords[0], coords[1], :] = 0
    return noisy

Y = add_salt_and_pepper_noise(A, 0.05)

# Mostrar la imagen con ruido sal y pimienta
plt.subplot(1, 3, 2)
plt.imshow(Y)
plt.title('Imagen con ruido sal y pimienta')
plt.axis('off')

# Eliminar ruido con un filtro de mediana
Z = cv2.medianBlur(Y, 3)

# Mostrar la imagen limpiada
plt.subplot(1, 3, 3)
plt.imshow(Z)
plt.title('Imagen limpiada')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen original
A = cv2.imread('imagen1.jpg')
A_rgb = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB para mostrar correctamente con Matplotlib

# Mostrar la imagen original
plt.subplot(1, 2, 1)
plt.imshow(A_rgb)
plt.title('Imagen original')
plt.axis('off')

# Obtener las dimensiones de la imagen
(h, w) = A_rgb.shape[:2]

# Calcular el centro de la imagen
center = (w // 2, h // 2)

# Generar la matriz de transformaci√≥n para rotar la imagen 45 grados
M = cv2.getRotationMatrix2D(center, 45, 1.0)

# Aplicar la rotaci√≥n
B = cv2.warpAffine(A_rgb, M, (w, h))

# Mostrar la imagen rotada
plt.subplot(1, 2, 2)
plt.imshow(B)
plt.title('Imagen rotada')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Leer la imagen original
A = cv2.imread('imagen3.jpg')
A_rgb = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB para mostrar correctamente con Matplotlib

# Mostrar la imagen original
plt.subplot(1, 2, 1)
plt.imshow(A_rgb)
plt.title('Imagen original')
plt.axis('off')

# Convertir la imagen a escala de grises
A_gray = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)

# Aplicar la detecci√≥n de bordes usando el algoritmo Canny
B = cv2.Canny(A_gray, 100, 200)

# Mostrar la imagen con bordes resaltados
plt.subplot(1, 2, 2)
plt.imshow(B, cmap='gray')
plt.title('Imagen con bordes resaltados')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Leer la imagen original
A = cv2.imread('imagen3.jpg')
A_rgb = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB para mostrar correctamente con Matplotlib

# Convertir la imagen a escala de grises
A_gray = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)

# Calcular el umbral autom√°ticamente usando el m√©todo de Otsu
_, B = cv2.threshold(A_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Mostrar la imagen original
plt.subplot(1, 2, 1)
plt.imshow(A_rgb)
plt.title('Imagen original')
plt.axis('off')

# Mostrar la imagen segmentada
plt.subplot(1, 2, 2)
plt.imshow(B, cmap='gray')
plt.title('Imagen segmentada')
plt.axis('off')

plt.show()
```
